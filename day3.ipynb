```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, InputLayer
```

```python
import seaborn as sns
iris = sns.load_dataset("iris")
iris.head()
```

```python
from sklearn.model_selection import train_test_split
X = iris.iloc[:, :-1]
y = iris.iloc[:, -1]
train_X, test_X, train_y, test_y = train_test_split(X, y) # 75:25
print(train_X.shape)
```

```python
model = Sequential()
model.add(InputLayer(4)) # 입력 변수의 개수를 지정
model.add(Dense(50, activation='sigmoid'))
model.add(Dense(30, activation='sigmoid'))
model.add(Dense(3, activation='softmax')) # 출력층의 뉴런의 개수: 회귀는 1개, 분류는 클래스 라벨의 개수
model.summary()
```

```python
# 훈련을 정의
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy',
              metrics=['accuracy']) # 평가방법을 accuracy로 수정해주세요.
```

```python
model.fit(train_X, train_y, epochs=200)
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, concatenate

input_model = Input(shape=(4,))
dense1 = Dense(50, activation='relu')(input_model)
dense2 = Dense(80, activation='relu')(input_model)
dense3 = Dense(30, activation='relu')(input_model)
x = concatenate([dense1, dense2, dense3])
output_model = Dense(3, activation='softmax')(x)
model = Model(inputs=input_model, outputs=output_model)
model.summary()
```

```python
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])
```

```python
model.fit(train_X, train_y, epochs=100)
```
